#### Importing the needful libraries:

```python
import pandas as pd
import os
```

#### Changing the path and then using list comprehension and concatination method to save the dataset in 'dataframe1' & 'dataframe2':

```python
path = '/Users/yudiudyan/Desktop/Projects/Project/Energy Data Analysis/Gen1/[IN-023C]/2018/2018-12'
os.chdir(path)
for a in os.listdir():
    if a == ".DS_Store": continue
    else:
        path = '/Users/yudiudyan/Desktop/Projects/Project/Energy Data Analysis/Gen1/[IN-023C]/2018/2018-12'
        path = path+"/"+a
        os.chdir(path)
        dataframe1 = pd.concat([pd.read_csv(file,sep="\t") for file in os.listdir()])
```

```python
path = '/Users/yudiudyan/Desktop/Projects/Project/Energy Data Analysis/Gen1/[IN-023C]/2019/2019-01'
os.chdir(path)
for a in os.listdir():
    if a == ".DS_Store": continue
    else:
        path = '/Users/yudiudyan/Desktop/Projects/Project/Energy Data Analysis/Gen1/[IN-023C]/2019/2019-01'
        path = path+"/"+a
        os.chdir(path)
        dataframe2 = pd.concat([pd.read_csv(file,sep="\t") for file in os.listdir()])
```

#### Using concatination, rename and  drop methods from pandas for a cleaner represenation of dataset:

```python
dataframe = pd.concat([dataframe1,dataframe2])
dataframe = dataframe.rename(columns={'m3':'Timestamp','m2':'Substation_ID'})
dataframe = dataframe.drop(columns=["Tstamp"])
dataframe.head()
```

#### Checking the total count of the 'm4' column: 

```python
dataframe.describe()
```

#### Splitting the date from the 'Timestamp' column into 'Year' for the years and 'Year _Month' for the months and years:

```python
year,year_month = [],[]
for timestamp in dataframe['Timestamp']:
    split = timestamp.split("-")
    year.append(split[0])
    year_month.append(split[0]+"-"+split[1])
dataframe['Year'],dataframe['Year_Month'] = year , year_month
dataframe.head()
```

#### Dropping all columns with null values:

```python
null_value_columns = []
for column in dataframe:
    if dataframe[column].isna().sum() != 0:
        null_value_columns.append(column)
dataframe = dataframe.drop(columns=null_value_columns)
dataframe.head()
```
